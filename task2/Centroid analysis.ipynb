# Data analysis

```python
DATA_PATH = '/home/saad/data/rob535/'
```

```python
centroid_path = DATA_PATH + 'trainval/centroids.csv'
```

```python
count = 0
dims = {'x': 0, 'y': 0, 'z': 0}
with open(centroid_path, 'r') as f:
    for line in f:
        if count == 0:
            count += 1
            continue
        print(line.split(',')[-1])
        val = float(line.split(',')[-1])
        guid, idx, dim = line.split(',')[0].split('/')
        dims[dim] += val
        count += 1
```

```python
for dim in dims:
    dims[dim] /= (count / 3)
```
